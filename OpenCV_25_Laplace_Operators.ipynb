```python
import cv2
import matplotlib.pyplot as plt
import numpy as  np
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('DATA/sudoku.jpg',0)
```

```python
display_img(img)
```

## Sobel Operators X gradient

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) #cv_64f is a depth value of output image, given value is highest precision.
# sobel(source,depth_Value,order of the derivative in x direction,order of the derivative in y direction,kenerl size)
```

```python
display_img(sobelx)
```

## Sobel Operators Y gradient

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5) #cv_64f is a depth value of output image, given value is highest precision.
# sobel(source,depth_Value,order of the derivative in x direction,order of the derivative in y direction,kenerl size)
```

```python
display_img(sobely)
```

## Laplace Derivative Operator

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```
