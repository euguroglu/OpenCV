```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Defining convenient function to read image quickly

```python
def load_img():
    img = cv2.imread('DATA/bricks.jpg').astype(np.float32)/255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

## Defining convenient function to display image quickly and larger scale

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

## Gamma Correction Method

Set gamma lower than 1 to get picture brigter

```python
gamma = 1/4
```

```python
result = np.power(i,gamma)
display_img(result)
```

Set gamma higher than 1 to get picture darker

```python
gamma = 2
```

```python
result = np.power(i,gamma)
display_img(result)
```
